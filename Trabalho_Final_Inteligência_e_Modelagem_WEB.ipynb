# Levantamento de dados - Queries

```python
import pandas as pd
import json
def sparql_result_to_df(sparql_result):
    # converts SPARQL SELECT results to Pandas dataframe
    
    cols = sparql_result.bindings[0].keys()

    out = []
    for row in sparql_result.bindings:
        item = []
        for c in cols:
            item.append(row.get(c, {}))
        out.append(item)
    return pd.DataFrame(out, columns=cols)
```

O número total de cursos é equivalente ao do ano de 2017, conforme informado na fonte da base de dados. É possível que os dados de 2018 e 2019 não estejam na base RDF.

```python
result = g.query("""
SELECT ?entity ?entity_label 
       (COUNT(?entity_label) AS ?entity_label_count)
WHERE {
    ?entity rdfs:label ?entity_label .
}
GROUP BY ?entity
ORDER BY DESC(?entity_label_count)

""")
df = sparql_result_to_df(result)
df
```

Quantidade de cursos com o mesmo nome

```python
result2 = g.query("""
SELECT ?nome_curso 
       (COUNT(?nome_curso) AS ?quantidade)
WHERE {
    ?entity skos:prefLabel ?nome_curso .
}
GROUP BY ?nome_curso
ORDER BY DESC(?quantidade)

""")
df2 = sparql_result_to_df(result2)
df2
```

Relação dos nomes dos cursos e notas obtidas (Conceito) em ordem decrescente.

```python
result3 = g.query("""
SELECT ?Curso ?Conceito 
WHERE {
    ?entity capes:conceitoCurso ?Conceito .
    ?entity rdfs:label ?Curso .
}
GROUP BY ?Curso
ORDER BY DESC(?Conceito)

""")
df3 = sparql_result_to_df(result3)
df3
```

Conceito médio por área de conhecimento em ordem decrescente. Visto que apenas os cursos possuem o nome registrado na base de dados, foi possível somente listar as áreas pelo seu ID.

```python
result4 = g.query("""
SELECT ?Area 
       (AVG(?Conceito) AS ?Conceito_medio)
WHERE {
    ?id capes:conceitoCurso ?Conceito .
    ?id capes:areaConhecimento ?Area .
}
GROUP BY ?Area
ORDER BY DESC (?Conceito_medio)
""")
df4 = sparql_result_to_df(result4)
df4.head(100)
```

Busca por algum curso específico filtrando por algum termo específico no nome. Nesse exemplo, foram buscados os cursos que possuem a palavra "operacional" no seu nome sem fazer distinção de letras maiúsculas ou minúsculas.

```python
result5 = g.query("""
SELECT ?Curso_especifico 
WHERE {
    ?id rdfs:label ?Curso_especifico .
    FILTER regex(?Curso_especifico, "operacional", "i")
}
GROUP BY ?Curso_especifico
""")
df5 = sparql_result_to_df(result5)
df5
```

Relação de quantos cursos por área de conhecimento.

```python
result6 = g.query("""
SELECT ?Area 
       (COUNT(?Area) AS ?area_count)
WHERE {
    ?entity capes:areaConhecimento ?Area .
}
GROUP BY ?Area
ORDER BY DESC(?area_count)

""")
df6 = sparql_result_to_df(result6)
df6
```

Listagem em ordem decrescente filtrando os cursos por nome. Nesse exemplo, os cursos foram restritos aos que possuem Engenharia de Produção no seu nome.

```python
result7 = g.query("""
SELECT ?Curso_especifico ?Conceito
WHERE {
    ?id capes:conceitoCurso ?Conceito .
    ?id rdfs:label ?Curso_especifico .
    FILTER regex(?Curso_especifico, "Engenharia de Prod", "i")
}
GROUP BY ?Curso_especifico
ORDER BY DESC (?Conceito)
""")
df7 = sparql_result_to_df(result7)
df7
```

No exemplo abaixo, foram consideradas as áreas de formação dos integrantes do grupo (Administração, Engenharia de Produção e Ciência da Computação), filtrados os cursos que possuem nomes próximos e ordenados de forma decrescente com base no seu conceito.

```python
result8 = g.query("""
SELECT ?Curso_especifico ?Conceito
WHERE {
    ?id rdfs:label ?Curso_especifico .
    ?id capes:conceitoCurso ?Conceito 
    FILTER (regex(?Curso_especifico, "Administra", "i") || 
    regex(?Curso_especifico, "de Produ", "i") || 
    regex(?Curso_especifico, "computa", "i"))
}
GROUP BY ?Curso_especifico
ORDER BY DESC(?Conceito)
""")
df8 = sparql_result_to_df(result8)
df8
```

Como alguns cursos de mestrado e doutorado possuem essas palavras no seu nome (ex.: Curso de mestrado de...), foi realizada uma contagem estimando aproximadamente quantos cursos de mestrado há na base. A mesma pesquisa pode ser feita para o caso do doutorado, ajustando apenas o filtro.

```python
result9 = g.query("""
SELECT (COUNT(?Mest) AS ?Cursos_de_Mestrado)
WHERE {
    ?id rdfs:label ?Mest .
    FILTER (regex(?Mest, "mestrado", "i")) .
}

""")
df9 = sparql_result_to_df(result9)
df9
```
